8. Design a function item_size_info that takes a NumPy array as input and returns the item size and the total size in bytes.

````python
import numpy as np
def item_size_info(arr):
    item_size = arr.itemsize
    total_size = arr.nbytes
    return item_size, total_size

arr_ex = np.array([1, 2, 3], dtype=np.int32) # int32 is 4 bytes
item_s, total_s = item_size_info(arr_ex)
print(f"Item Size (bytes): {item_s}, Total Size (bytes): {total_s}")
# Output: Item Size (bytes): 4, Total Size (bytes): 12 (4 * 3 elements)
````

18. Write a function that takes two NumPy arrays and concatenates them along a specified axis.

````python
import numpy as np
def concatenate_arrays(arr1, arr2, axis=0):
    return np.concatenate((arr1, arr2), axis=axis)

arr_a = np.array([[1, 2], [3, 4]])
arr_b = np.array([[5, 6], [7, 8]])

# Concatenate along axis=0 (rows)
concat_rows = concatenate_arrays(arr_a, arr_b, axis=0)
print("Concatenated along axis 0:\n", concat_rows)
# Output: [[1 2]
#          [3 4]
#          [5 6]
#          [7 8]]

# Concatenate along axis=1 (columns)
concat_cols = concatenate_arrays(arr_a, arr_b, axis=1)
print("Concatenated along axis 1:\n", concat_cols)
# Output: [[1 2 5 6]
#          [3 4 7 8]]
````

25. Write a function that takes a Pandas DataFrame and replaces all negative values in a specific column with zeros. Use NumPy operations within the Pandas DataFrame.

````python
import numpy as np
def insert_and_delete(arr, ins_idx, ins_val, del_idx):
    # 1. Insert elements
    arr_inserted = np.insert(arr, ins_idx, ins_val)

    # 2. Delete elements from the inserted array
    arr_final = np.delete(arr_inserted, del_idx)
    return arr_inserted, arr_final

original_array = np.array([1, 2, 3, 4, 5])
indices_to_insert = [2, 4]
values_to_insert = [10, 11]
indices_to_delete = [1, 3]

arr_ins, arr_del = insert_and_delete(original_array, indices_to_insert, values_to_insert, indices_to_delete)

print(f"Array after insertion: {arr_ins}")
print(f"Array after deletion: {arr_del}")
````

34. Write a NumPy function that returns elements from an array where both two conditions are satisfied using boolean indexing.

````python
import numpy as np
def filter_by_two_conditions(arr, cond1, cond2):
    return arr[(cond1) & (cond2)]

arr = np.array([10, 20, 30, 40, 50])
# Condition 1: element > 20, Condition 2: element <= 40
cond1 = (arr > 20)
cond2 = (arr <= 40)

filtered = filter_by_two_conditions(arr, cond1, cond2)
print(f"Elements > 20 AND <= 40: {filtered}")
````
